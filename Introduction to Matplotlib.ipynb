# matplotlib example worflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline

import matplotlib.pyplot as plt
# 1. Prepare data
x = [1, 2, 3, 4]
y =[11, 22, 33, 44]

# 2 Setup plot
fig, ax = plt.subplots(figsize = (10, 10))

# 3 plot the data
ax.plot(x, y)

# 4 Customize plot
ax.set(title = 'Simple Plot', xlabel = 'x-axis', ylabel = 'y-axis')

# 5 Save & show the whole figure
fig.savefig('images/sample-plot.png')
```

### Two Options for subplot

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize= (10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_store.keys(), nut_butter_store.values());
ax4.hist(np.random.randn(1000))
```

```python
# Subplot Option2
fig, ax = plt.subplots(nrows=2, ncols=2,figsize=(10,5))

#Plot for each different index
ax[0, 0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_store.keys(), nut_butter_store.values())
ax[1,1].hist(np.random.randn(1000));
```

### which one should you use pyplot or matplotlib oo methods ?
* When plotting something quickly use pyplot
* When plotting something more advanced use oo method 

```python
# Create a subset dataframe
over_50 = heart_disease[heart_disease.age > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind = 'scatter', x = 'age', y = 'chol', c = 'target');
```

```python
## OO method mixed with pyplot
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter', x = 'age', y = 'chol', ax= ax, c= 'target' );
# ax.set_xlim(45,100);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

#Plot the data
scatter = ax.scatter(x = over_50['age'], y = over_50['chol'], c= over_50['target'] )

#Customize the plot
ax.set(title = 'Heart disease and Cholestrol Levels', xlabel = 'Age', ylabel = 'Cholestrol Levels');

#Add a legend
ax.legend(*scatter.legend_elements(), title = 'Target');

# Add a horizontol line
ax.axhline(over_50['chol'].mean(), linestyle = '--');
```

```python
over_50.head(0)
```

```python
# Subplot of col, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex = True)

scatter1 = ax0.scatter(x = over_50['age'], y = over_50['chol'], c= over_50['target'])
ax0.set(title = 'Heart Disease and Cholestrol Levels', ylabel = 'Cholestrol')

#Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = 'Target')

# Add a mean line
ax0.axhline(over_50['chol'].mean(), linestyle = '--')

# Do a subplot for thalach column
scatter2 = ax1.scatter(x = over_50['age'], y = over_50['thalach'], c = over_50['target'])

# Customize the second plot
ax1.set(title = 'Heart disease and Max Heart Rate', xlabel = 'Age', ylabel = 'Max Heart levels')
# Add a legend for target variable
ax1.legend(*scatter.legend_elements(), title = 'Target')

# Add a mean line for average thalach levels
ax1.axhline(over_50['thalach'].mean(), linestyle = '--');

#Add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold')
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

#Plot the data
scatter = ax.scatter(x = over_50['age'], y = over_50['chol'], c= over_50['target'] ,cmap = 'winter')

#Customize the plot
ax.set(title = 'Heart disease and Cholestrol Levels', xlabel = 'Age', ylabel = 'Cholestrol Levels');

#Add a legend
ax.legend(*scatter.legend_elements(), title = 'Target');

# Add a horizontol line
ax.axhline(over_50['chol'].mean(), linestyle = '--');
```

```python
# Subplot of col, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex = True)

scatter1 = ax0.scatter(x = over_50['age'], y = over_50['chol'], c= over_50['target'], cmap = 'winter')
ax0.set(title = 'Heart Disease and Cholestrol Levels', ylabel = 'Cholestrol')
#Change the x limits:
ax0.set_xlim([50, 90])
#Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = 'Target')

# Add a mean line
ax0.axhline(over_50['chol'].mean(), linestyle = '--')

# Do a subplot for thalach column
scatter2 = ax1.scatter(x = over_50['age'], y = over_50['thalach'], c = over_50['target'], cmap = 'winter')

# Customize the second plot
ax1.set(title = 'Heart disease and Max Heart Rate', xlabel = 'Age', ylabel = 'Max Heart levels')

#Change ax1 x and y limits:
ax1.set_xlim([50, 90])
ax1.set_ylim([60, 200])

# Add a legend for target variable
ax1.legend(*scatter.legend_elements(), title = 'Target')

# Add a mean line for average thalach levels
ax1.axhline(over_50['thalach'].mean(), linestyle = '--');

#Add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold')
```
